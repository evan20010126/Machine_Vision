## 執行下面程式測試安裝結果

下面程式使用Res50模型辨識下面老虎的影像。
<div>
    <img src="https://static.bangkokpost.com/media/content/20190729/c1_1720911_190729141725.jpg" width="400">
</div>

```python
import matplotlib.pyplot as plt
# display plots in this notebook
%matplotlib inline 
import cv2
import numpy as np
from tensorflow.keras.applications.resnet import ResNet50
from tensorflow.keras.applications.resnet import preprocess_input, decode_predictions
from tensorflow.keras.utils import plot_model
from tensorflow.keras.preprocessing import image
from IPython.display import display

def show_model(model,figsize=(20,20)):
    a = plot_model(model, model.name +'.png', show_shapes=True, dpi=64)
    display(a)
    return
```

```python
model = ResNet50() # create a deep CNN
#model.summary()   # display the network architecture
try:    
# display and save the figure of the the network architecture into "model.name".png.    
    show_model(model,figsize=(60,300))    
except:
    pass
```

**下載測試影像**

```python
url='https://as.chdev.tw/web/article/b/0/4/b1814323-0790-4b70-a3b1-6cbc87d37d1f/A0951614.jpg'
name   =url.split('/')[-1]
srcfile= tf.keras.utils.get_file(name,origin=url)
test_opencv = True

if test_opencv:
    img1 = cv2.resize(cv2.imread(srcfile),(224,224))[:,:,::-1].astype('float32')
else:
    img1 = image.load_img(srcfile,target_size=(224,224))
    img1 = image.img_to_array(img1)

plt.figure()
plt.imshow(img1[:,:,:]/255)
plt.axis('off')
plt.show()
```

**影像辨識**

```python
preds = model.predict(preprocess_input(np.expand_dims(img1.copy(), axis= 0)))
print(preds.shape, np.argmax(preds.ravel()))
P = decode_predictions(preds)

for (i, (imagenetID, label, prob)) in enumerate(P[0]):
    print("{}. {}({}): {:.2f}%".format(i + 1, label, imagenetID, prob * 100))
```

**測試iris資料集**

```python
from scipy import stats
from sklearn import neighbors
from sklearn import datasets
from sklearn.model_selection import StratifiedKFold, cross_val_score, GridSearchCV
from sklearn import svm

import numpy as np

C_param_range     = np.logspace(-3,3,7)
gamma_param_range = np.logspace(-3,2,6)

# load the test dataset
iris = datasets.load_iris()

# inner cross-validation for hyper-parameter tuning
A_param_grid = [{'C': C_param_range, 'kernel': ['linear']},{'C': C_param_range, 'kernel': ['rbf'],'gamma':gamma_param_range}]
B_param_grid = [{'n_neighbors':[1,3,5,7,9], 'weights':['uniform','distance']}]

A_gs = GridSearchCV(estimator=svm.SVC(),param_grid = A_param_grid, scoring = 'accuracy', cv=5, n_jobs=-1)
B_gs = GridSearchCV(estimator=neighbors.KNeighborsClassifier(),param_grid = B_param_grid, scoring = 'accuracy', cv=5, n_jobs=-1)

# outer cross-validation for estimating the accuracy of the classifier
kfold = StratifiedKFold(n_splits=10, shuffle=True, random_state=2)
A_scores = cross_val_score(A_gs, iris.data, iris.target, scoring='accuracy',cv = kfold)
B_scores = cross_val_score(B_gs, iris.data, iris.target, scoring='accuracy',cv = kfold)

for idx, (a,b) in enumerate(zip(A_scores,B_scores)):
    print('{:2d} SVC:{:.2f} vs kNN:{:.2f}'.format(idx,a,b))
print('average:{:.2f}+-{:.2f}, {:.2f}+-{:.2f}'.format(np.mean(A_scores),np.std(A_scores),np.mean(B_scores),np.std(B_scores)))

# this is a two-sided test that the null hypothesis that the expected value (mean) of a sample of 
# independent observations a is equal to the given population mean
t,pvalue=stats.ttest_rel(A_scores, B_scores)
if pvalue<= 0.05:
     print('the difference in accuracy between classifiers A and B is significant p-value={:.3f}'.format(pvalue))
else:
     print('the difference in accuracy between classifiers A and B is not significant p-value={:.3f}'.format(pvalue))
```

## 建立虛擬環境
針對不同的Python模組或專案，可建立各自虛擬環境，以免因版本問題互相衝突。在Anaconda Prompt視窗中下達指令。
<div>
    <img src="https://tronclass.ntou.edu.tw/api/uploads/1697659/in-rich-content?created_at=2021-02-21T08:38:43Z" width="400">
</div>

+ conda create -n env
+ conda create python=3.7 -n env

<a href="https://docs.conda.io/projects/conda/en/latest/user-guide/tasks/manage-environments.html" target="_blank" title="conda">conda虛擬環境操作指令</a>

#### Activate a virtual environment
+ conda activate env

#### Deactivate a virtual environment
+ conda deactivate

#### List virtual environment
+ conda env list

#### Rename a virtual environment
以下面兩個指令完成改虛擬環境名稱
+ conda create --name new_name --clone old_name
+ conda remove –-name old_name --all

#### Create a virtual environment in a specific directory
+ conda create --prefix  /new_directory/env_name python=3
+ This command will create the environment named /new_directory/env_name which resides in /new_directory instead of the default 	e.g., E:\ProgramData\Anaconda3\envs
